```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('train_eda.csv')
```

```python
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar', stacked=True)
plt.show()
```

```python
table = pd.crosstab(df['Sex'], df['Pclass'])

plt.imshow(table, cmap='Blues')
plt.xticks(range(len(table.columns)), table.columns)
plt.yticks(range(len(table.index)), table.index)
plt.colorbar(label="Count")
plt.xlabel("Pclass")
plt.ylabel("Sex")
plt.show()
```

```python
data = [
    df[df['Pclass']==1]['Age'].dropna(),
    df[df['Pclass']==2]['Age'].dropna(),
    df[df['Pclass']==3]['Age'].dropna()
]

plt.boxplot(data, labels=[1,2,3])
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python

import matplotlib.pyplot as plt



survival_counts = df['Survived'].value_counts()

plt.figure()
plt.bar(['Not Survived','Survived'], survival_counts.values)
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.show()
```

```python


#survival_by_class

class_survival = df.groupby('Pclass')['Survived'].sum()

plt.figure()
plt.bar(class_survival.index.astype(str), class_survival.values)
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.show()

```
